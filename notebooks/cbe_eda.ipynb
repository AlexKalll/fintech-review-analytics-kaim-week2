```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style
sns.set(style="whitegrid")
```

```python
file_path = "../data/processed/cbe_clean_reviews.csv"
df = pd.read_csv(file_path)
print(f"Loaded {len(df)} cleaned reviews.")
df.head()
```

```python
df.info()
df.describe()
df['rating'].value_counts().sort_index()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='rating', palette='viridis')
plt.title("CBE App Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['sentiment'], bins=30, kde=True, color='blue')
plt.title("Sentiment Score Distribution (CBE)")
plt.xlabel("Sentiment Polarity")
plt.ylabel("Frequency")
plt.show()
```

```python
for rating in [1, 3, 5]:
    print(f"\nSample reviews with rating {rating}:\n")
    print(df[df['rating'] == rating]['review'].sample(3, random_state=42).values)
```
